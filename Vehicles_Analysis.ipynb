### Feature Engineering

- Feature engineering is a creative process. TO adding new features,need to provide your model with more information to find patterns and make predictions. Always test and validate with new features with the model to ensure they're improving performance.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Example: Check the column names to find appropriate numerical features
print(df.columns)

# Automatically detect numerical columns
numerical_features = df.select_dtypes(include=['float64', 'int64']).columns

# Step 1: Create New Features (only using numerical columns)
# Polynomial Feature example (use the first numerical column)
if len(numerical_features) > 0:
    df['numerical_feature_squared'] = df[numerical_features[0]] ** 2  # Square the first numerical feature

# Rolling statistics example (5-period rolling mean and std deviation)
if len(numerical_features) > 1:
    df['rolling_mean'] = df[numerical_features[1]].rolling(window=5).mean()  # 5-period rolling mean
    df['rolling_std'] = df[numerical_features[1]].rolling(window=5).std()  # 5-period rolling std deviation

# Interaction feature example (product of two numerical columns)
if len(numerical_features) > 2:
    df['interaction_feature'] = df[numerical_features[0]] * df[numerical_features[2]]  # Interaction feature

# Step 2: Standardization / Normalization
# Standardization (z-score)
scaler_standard = StandardScaler()
df[numerical_features] = scaler_standard.fit_transform(df[numerical_features])

# Normalization (min-max scaling)
scaler_minmax = MinMaxScaler()
df[numerical_features] = scaler_minmax.fit_transform(df[numerical_features])

# Step 3: One-Hot Encoding for Categorical Features (if available)
df = pd.get_dummies(df, drop_first=True)

# Output the first few rows to check the changes
print(df.head())
```
